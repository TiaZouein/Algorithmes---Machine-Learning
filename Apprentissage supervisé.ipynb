# Apprentissage supervisé

De toute évidence, la machine apprendra plus rapidement avec un enseignant, c'est pourquoi l'aprentissage supervisé est plus couramment utilisé dans les tâches réelles.

Les problèmes d'apprentissage supervisé sont catégorisés en problèmes de "régression" et de "classification". Dans un problème de régression, nous essayons de prédire des résultats dans une sortie continue, ce qui signifie que nous essayons de faire correspondre des variables d'entrée à une fonction continue. Dans un problème de classification, nous essayons plutôt de prédire des résultats dans une sortie discrète. En d'autres termes, nous essayons de faire correspondre des variables d'entrée à des catégories discrètes.

1- Classification

L'apprentissage supervisé en classification est une technique d'apprentissage automatique (machine learning) où un modèle est formé pour prédire une catégorie ou une classe (étiquette) à partir de données d'entraînement préalablement étiquetées. Le but de l'apprentissage supervisé en classification est de créer un modèle capable de généraliser à de nouvelles données non étiquetées et d'attribuer correctement des étiquettes aux nouvelles observations.

2- Régression

La régression est l'une des techniques courantes en apprentissage automatique (machine learning) utilisée pour prédire une valeur numérique (variable dépendante) à partir de données d'entrée. Elle s'applique lorsque vous souhaitez modéliser la relation entre des variables indépendantes (caractéristiques) et une variable dépendante (généralement continue) afin de faire des prédictions ou d'analyser cette relation.

Utilisations actuelles :

Prévisions des prix des actions
Analyse de la demande et du volume de ventes
Diagnostic médical
Corrélations temporelles de tout type
Les algorithmes populaires sont les régressions linéaires et polynomiales.
